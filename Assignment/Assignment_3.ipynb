```
#a) Create HR_Employee Pysaprk dataFrame.
hr_employee = spark.read.csv("dbfs:/FileStore/tables/HR_Employee.csv", header=True, inferSchema=True)
```

```
#b) Show printSchema &amp; show 10 rows of Dataset
hr_employee.printSchema()
hr_employee.show(10, False)
```

```
#c) Most employee is working in which department
hr_employee.createTempView("employee_table")
```

```
%sql
SELECT * from employee_table where LIMIT 10
```

```
spark.sql("SELECT Department,count(*) as total_no_employess from employee_table GROUP BY Department order by total_no_employess desc").show(truncate=False)
```

```
#d) Highest number of job roles
spark.sql("SELECT JobRole,count(*) as high_job_role from employee_table GROUP BY JobRole order by high_job_role desc").show(1,truncate=False)
```

```
#e) Which gender have higher strength as workforce?
spark.sql("SELECT Gender,count(*) as workforce_count from employee_table GROUP BY Gender order by workforce_count desc").show(truncate=False)
```

```
#f) Compare the marital status of employee and find the most frequent status.
spark.sql("SELECT MaritalStatus,count(*) as marital_status_count from employee_table GROUP BY MaritalStatus order by marital_status_count desc").show(truncate=False)
```

```
#g) Mostly hired employee have qualification
spark.sql("SELECT Education,count(*) as qualification_details from employee_table GROUP BY Education order by qualification_details desc").show(1,truncate=False)
```

```
#h) Find the count of employee from which education fields
spark.sql("SELECT Education,count(*) as qualification_details from employee_table GROUP BY Education order by qualification_details desc").show(truncate=False)
```

```
#i) What is the job satisfaction level of employee?
spark.sql("SELECT JobSatisfaction,count(*) as satisfaction_level from employee_table GROUP BY JobSatisfaction order by satisfaction_level desc").show(truncate=False)
```

```
#j) Does most of employee do overtime: Yes or No?
spark.sql("SELECT JobSatisfaction,count(*) as satisfaction_level from employee_table GROUP BY JobSatisfaction order by satisfaction_level desc").show(truncate=False)
```

```
#k) Find Min &amp; Max Salaried employees.
spark.sql("SELECT min(Income) as minimum_salary,max(Income) as max_salary FROM employee_table").show()
```

```
#l) Does most of the employee do business travel? Find of the employees counts for each category
spark.sql("SELECT BusinessTravel,count(*) as travel_frequency from employee_table GROUP BY BusinessTravel order by travel_frequency desc").show(truncate=False)
```

```
#m) Find the AVG salary of graduate employee.
spark.sql("SELECT avg(income) as avg_grad_income from employee_table where education not in ('below college')").show()
```

```
#n) Find the employee qualification receiving salary lower than equal to avg. salary of all employee.
spark.sql("SELECT education,income from employee_table where income <= ( SELECT avg(income) as avg_grad_income from employee_table)").show()
```

```
#o) When does the employee have highest chance of promotion in terms of working year?
spark.sql("SELECT count(*) as promoted_employees,(Workex - YearsSinceLastPromotion) as year_experience from employee_table group by (Workex - YearsSinceLastPromotion) order by promoted_employees desc").show()
```

```
#p) Highest attrition is in which department?
spark.sql("SELECT department,count(*) as total_attrition from employee_table where attrition='Yes' GROUP BY department ORDER BY total_attrition desc").show(1,truncate=False)
```

```
#q) Show marital status of Person having highest attrition rate
spark.sql("SELECT MaritalStatus,count(*) as total_attrition from employee_table where attrition='Yes' GROUP BY MaritalStatus ORDER BY total_attrition desc").show(truncate=False)
```
